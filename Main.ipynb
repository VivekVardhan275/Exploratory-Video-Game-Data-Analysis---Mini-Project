Importing Pyhton Data Analysis libraries and loading the data

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

Reading the data and displaying the first few rows

```python
vgsales_df = pd.read_csv('vgsales.csv')
```

```python
vgsales_df.head(15)
```

Describing the data

```python
vgsales_df.describe()
```

Information about the data

```python
vgsales_df.info();
```

Information about the null values

```python
print(vgsales_df.isnull().sum())
```

Cleaning the null values and checking the data

```python
vgsales_df = vgsales_df.dropna();
```

```python
vgsales_df.info();
```

```python
vgsales_df.isnull().sum()
```

Total Games

```python
total_games = vgsales_df.Name.value_counts()
total_games
```

Total Platforms

```python
total_platforms = vgsales_df.Platform.value_counts()
total_platforms
```

Total Genres

```python
total_generes = vgsales_df.Genre.value_counts()
total_generes
```

Visualizing the NA sales of the games with respect to year

```python
plt.figure(figsize=(15,5))
x_axis = vgsales_df['Year'];
y_axis = vgsales_df['NA_Sales'];
plt.title('North America Sales')
sns.lineplot(x=x_axis, y=y_axis, data=vgsales_df , color='blue')
```

Visualizing the EU sales of the games with respect to year

```python
plt.figure(figsize=(15,5))
x_axis = vgsales_df['Year'];
y_axis = vgsales_df['EU_Sales'];
plt.title('Europe Sales')
plt.xlabel('Year')
plt.ylabel('EU Sales')
sns.lineplot(x=x_axis, y=y_axis, data=vgsales_df)
```

Visualizing the Asian sales of the games with respect to year

```python
plt.figure(figsize=(15,5))
x_axis = vgsales_df['Year'];
y_axis = vgsales_df['JP_Sales'];
plt.title('Japan Sales')
plt.xlabel('Year')
plt.ylabel('JP Sales')
sns.lineplot(x=x_axis, y=y_axis, data=vgsales_df,color = 'green')
```

Visualizing the Global sales of the games with respect to year

```python
plt.figure(figsize=(15,5))
x_axis = vgsales_df['Year'];
y_axis = vgsales_df['Global_Sales'];
plt.title('Global Sales')
plt.xlabel('Year')
plt.ylabel('Global Sales')
sns.lineplot(x=x_axis, y=y_axis, data=vgsales_df,color = 'violet')
```

Top 3 games Global sales

```python
plt.figure(figsize=(7,7))
x_axis = vgsales_df['Name'][0:3];
y_axis = vgsales_df['Global_Sales'][0:3];
plt.title('Top 3 Games Global Sales')
plt.xlabel('Name')
plt.ylabel('Global Sales')
sns.barplot(x=x_axis, y=y_axis, data=vgsales_df ,hue =x_axis)
```

Most Games released in the year and visualizing the data

```python
most_games = vgsales_df['Year'].value_counts()
most_games
```

```python
mostgames_df = vgsales_df[vgsales_df['Year'] == 2009]
mostgames_df
```

```python
#Total games in 2009
mostgames_df.value_counts().sum()
```

```python
#visualizing games sales in 2009
sales = mostgames_df[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales' , 'Global_Sales']]
plt.figure(figsize=(7,7))
sns.barplot(sales )
```

Visualizing the data based on the Genre

```python
plt.figure(figsize=(15,5))
sns.barplot(x='Genre', y='Global_Sales', data=vgsales_df , hue='Genre')
```

Data between publishers and the sales

```python
plt.figure(figsize=(15,5))
publisher_df   = vgsales_df['Publisher'].value_counts()
x_axis = publisher_df[0:15].index
y_axis = publisher_df[0:15].values
sns.barplot(x=x_axis, y=y_axis , hue=x_axis)
```

Data According to Publishers

```python

Sony_Pub = vgsales_df[(vgsales_df['Publisher'] == 'Sony Computer Entertainment')]
Sony_Pub.value_counts().sum()
```

```python
plt.figure(figsize=(12, 6))
sns.scatterplot(Sony_Pub , x = 'Year', y = 'Global_Sales' , hue = 'Platform')
plt.title('Sony Computer Entertainment [Publisher]')
plt.xlabel('Year')
plt.ylabel('Global Sales')
```

Global sales above 30

```python
G_50 = vgsales_df[vgsales_df['Global_Sales'] > 30]
G_50
x_axis = G_50['Name']
y_axis = G_50['Global_Sales']
plt.figure(figsize=(15,5))
sns.barplot(x=x_axis, y=y_axis , hue=x_axis)
plt.title('Games with Global Sales > 30')
```

Sales distribution by platform
